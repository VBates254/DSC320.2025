```python
# Week 9 Problem 2 - Punnett Squares but with Math and Python 
# Parts 1-4
import pandas as pd

# Load the dataset
df = pd.read_csv('C:/Users/vampr/Desktop/Week 9 Data/survey_data.csv')

# Total number of students
total_students = len(df)

# 1. Probability of brown hair
p_brown_hair = (df['hair_color'] == 'brown').sum() / total_students

# 2. Probability of blue eyes
p_blue_eyes = (df['eye_color'] == 'blue').sum() / total_students

# 3. Probability of blue eyes given brown hair (P(Blue Eyes | Brown Hair))
brown_hair_df = df[df['hair_color'] == 'brown']
p_blue_eyes_given_brown_hair = (brown_hair_df['eye_color'] == 'blue').sum() / len(brown_hair_df)

# 4. Probability of brown hair given blue eyes (P(Brown Hair | Blue Eyes))
blue_eyes_df = df[df['eye_color'] == 'blue']
p_brown_hair_given_blue_eyes = (blue_eyes_df['hair_color'] == 'brown').sum() / len(blue_eyes_df)

print(f"P(Brown Hair): {p_brown_hair:.3f}")
print(f"P(Blue Eyes): {p_blue_eyes:.3f}")
print(f"P(Blue Eyes | Brown Hair): {p_blue_eyes_given_brown_hair:.3f}")
print(f"P(Brown Hair | Blue Eyes): {p_brown_hair_given_blue_eyes:.3f}")
```

```python
# Are brown hair and blue eyes independent? Let's make Python tell us
# P(Brown Hair and Blue Eyes)
p_both = len(df[(df['hair_color'] == 'brown') & (df['eye_color'] == 'blue')]) / total_students
p_independent = p_brown_hair * p_blue_eyes

print(f"P(Brown Hair ∩ Blue Eyes): {p_both:.3f}")
print(f"P(Brown Hair) * P(Blue Eyes): {p_independent:.3f}")
print("This suggests they are NOT independent, as they are not equal.")
print("Logic wants me to say of course they are but hey, I'm not a geneticist")
```

```python
# Bar Graphs
import matplotlib.pyplot as plt

# Set up subplots
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Hair color counts
df['hair_color'].value_counts().plot(kind='bar', ax=axs[0], color='brown')
axs[0].set_title('Hair Color Distribution')
axs[0].set_xlabel('Hair Color')
axs[0].set_ylabel('Count')

# Eye color counts
df['eye_color'].value_counts().plot(kind='bar', ax=axs[1], color='teal')
axs[1].set_title('Eye Color Distribution')
axs[1].set_xlabel('Eye Color')
axs[1].set_ylabel('Count')

plt.tight_layout()
plt.show()
```
