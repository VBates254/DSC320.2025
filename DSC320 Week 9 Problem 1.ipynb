```python
# Week 9
# Problem 1 - Mean and SD
import pandas as pd

# Load the dataset
df = pd.read_csv('C:/Users/vampr/Desktop/Week 9 Data/qb_stats.csv')

# Select only numerical columns
numeric_cols = [
    'games', 'att', 'cmp',
    'comp_pct', 'yds', 'ypa',
    'td', 'int', 'sack', 'loss',
    'rating'
]

# Calculate mean
mean_values = df[numeric_cols].mean()

# Calculate standard deviation
std_values = df[numeric_cols].std()

print("Mean of numerical columns:\n", mean_values)
print("\nStandard deviation of numerical columns:\n", std_values)
```

```python
# Histogram
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load the dataset
df = pd.read_csv('C:/Users/vampr/Desktop/Week 9 Data/qb_stats.csv')
plt.figure(figsize=(8, 6))
plt.hist(df['yds'], bins=15, color='gold', edgecolor='black') # who dat, dat dem saints
plt.title('Distribution of Passing Yards')                    # I dont even like football
plt.xlabel('Yards')                                           # I DO like stats tho. Kinda
plt.ylabel('Frequency')
plt.grid(False) # no grid lines
plt.show()

# Bowplot w/ OUtliers

plt.figure(figsize=(6, 5))
sns.boxplot(y=df['td'], color='purple')  # geux tigers - but specifically Mike the Tiger
plt.title('Boxplot of Touchdowns')
plt.ylabel('Touchdowns')
plt.grid(False)
plt.show()

#suspected outliers based on IQR
Q1 = df['td'].quantile(0.25)
Q3 = df['td'].quantile(0.75)
IQR = Q3 - Q1
outliers = df[(df['td'] < Q1 - 1.5 * IQR) | (df['td'] > Q3 + 1.5 * IQR)]

print("Outliers based on touchdowns:\n", outliers[['name', 'td']])
```
