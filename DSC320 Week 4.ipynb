```python
#Week 4
# 1 -Write a Python function to take in a mathematical function f, 
# two interval endpoints a and b, and the number of rectangles n, 
# and then outputs the right Riemann sum Rn. 
import math

# prep 
def right_riemann_sum(f, a, b, n): # f - function
    dx = (b - a) / n               # a - float/interval start
    total = 0                      # b - float/end
    for i in range(1, n + 1):      # n - integer/rectagles
        x_i = a + i * dx
        total += f(x_i)
    return total * dx

# function to integrate
def f(x):
    return math.sin(x)

# Reinmann sum
result = right_riemann_sum(f, 0, math.pi, 100)
print(f"Right Riemann Sum: {result}")
```

```python
# 2.a -Sketch a graph of f and the rectangles that make up the right Riemann sum 
# R4 on the interval [0, 2]. 
# I used matplotlib just as a personal preference.
# And because I don't have a good way to upload this if I do it by hand.

import numpy as np
import matplotlib.pyplot as plt

# function
def f(x):
    return 4 - x**2

a, b = 0, 2        # endpoints
n = 4              # rectangles
dx = (b - a) / n   # widths

x_right = np.linspace(a + dx, b, n)
y_right = f(x_right)

# curve
x_vals = np.linspace(a, b, 400)
y_vals = f(x_vals)
plt.plot(x_vals, y_vals, label='$f(x) = 4 - x^2$', color='blue')

# Plot rectangles
for i in range(n):
    plt.bar(x_right[i], y_right[i], width=dx, align='edge', 
            edgecolor='black', color='orange', alpha=0.6)

# labels
plt.title('Right Riemann Sum $R_4$ for $f(x) = 4 - x^2$ on [0,2]')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```
