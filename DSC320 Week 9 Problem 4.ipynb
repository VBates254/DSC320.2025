```python
# Week 9 Problem 4 - Random Data Woohoo
# 4.a - Make up your own probability mass function (pmf) for X. 
# Make sure it satisfies the requirements of a pmf.

# gotta have non-neg probablities that add up to 1
import numpy as np

# Define values and a made-up PMF
values = [1, 2, 3, 4, 5]
probabilities = [0.1, 0.2, 0.3, 0.3, 0.1]  # Sum is 1 

# 4.b &  4.c -Build a function that will generate 50 random X-values from the pmf you created in part (a) 
# and return the mean of these values.
def generate_sample_mean():
    sample = np.random.choice(values, size=50, p=probabilities)
    return np.mean(sample)

# now run it 1k times, (cha cha real smooth)
def run_simulation(n_runs=1000):
    means = [generate_sample_mean() for _ in range(n_runs)]
    return means

means_list = run_simulation()

# 4.d Histograms and Theoretical Shape
import matplotlib.pyplot as plt

plt.hist(means_list, bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram of Sample Means')
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')
plt.show()

print("Central Limit Theorem. Mostly bell-shaped, and we love that for it.")
```

```python
# 4.e - Empirical vs Theoretical
empirical_mean = np.mean(means_list)
empirical_std = np.std(means_list)

# Theoretical Mean and Std Dev of the PMF
theoretical_mean = np.dot(values, probabilities)
theoretical_std = np.sqrt(np.dot((np.array(values) - theoretical_mean)**2, probabilities))

# Std Dev of sample means = population std dev / sqrt(sample size)
theoretical_sample_std = theoretical_std / np.sqrt(50)

print(f"Empirical Mean: {empirical_mean:.4f}")
print(f"Empirical Std Dev: {empirical_std:.4f}")
print(f"Theoretical Mean: {theoretical_mean:.4f}")
print(f"Theoretical Std Dev of Sample Means: {theoretical_sample_std:.4f}")
```
