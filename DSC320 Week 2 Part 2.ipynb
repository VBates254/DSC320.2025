```python
# Week 2 
# 2.a - Make a histogram of the customer spending amounts. 
import pandas as pd
import matplotlib.pyplot as plt

file_path = r"C:\Users\vampr\Desktop\customer_spending.csv"  
df = pd.read_csv(file_path)

# Create histogram
plt.figure(figsize=(8, 5))
plt.hist(df["ann_spending"], bins=20, color="green", edgecolor="black")

# Add labels and title
plt.xlabel("Spending Amount ($)")
plt.ylabel("Frequency")
plt.title("Histogram of Customer Spending")

# Show the plot
plt.grid(False) #gridlines on this hurt my eyeballs so, no gridlines
plt.show()
```

```python
# 2.b -  Make a new data set that is a log transformation of the customer spending amounts. 
import pandas as pd
import numpy as np

file_path = r"C:\Users\vampr\Desktop\customer_spending.csv"
df = pd.read_csv(file_path)

# log transformation
df["Log_Spending"] = np.log(df["ann_spending"])

# Display the first few rows to verify
print(df.head())

# Save the transformed dataset for next part
df.to_csv(r"C:\Users\vampr\Desktop\customer_spending_log.csv", index=False)
print("Log-transformed dataset saved as 'customer_spending_log.csv'.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the log-transformed dataset from last part
file_path = r"C:\Users\vampr\Desktop\customer_spending_log.csv"
df = pd.read_csv(file_path)

# Check column names
print("Columns in dataset:", df.columns)

# Create histogram
plt.figure(figsize=(8, 5))
plt.hist(df["Log_Spending"], bins=20, color="green", edgecolor="black")

# Add labels and title
plt.xlabel("Spending Amount(by thousands of $) ")
plt.ylabel("Frequency")
plt.title("Histogram of Log-Transformed Customer Spending")

# Show the plot
plt.grid(False)
plt.show()
```
